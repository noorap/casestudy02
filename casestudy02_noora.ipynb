# 3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
# Average based on gender
avggen = data[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').mean()
avggen
```

Female students have higher average for reading and writing score while male students have higher average in maths score..

```python
# Average based on test preparation course
avgtest= data[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').mean()
avgtest
```

Students who completed test preparation course have higher average score for all scores

# 4. What do you think about the scoring variation for math, reading and writing based on
● Gender
● Test preparation course

```python
# Variation based on gender 
vargen= data[['gender', 'math score', 'reading score', 'writing score']].groupby('gender').std()
vargen
```

Male students have less variations for each of the scores.

```python
# Variation based on test preparation course
vartest= data[['test preparation course', 'math score', 'reading score', 'writing score']].groupby('test preparation course').std()
vartest
```

The variation for those who completed test preparation course is less compared to those who not done .
